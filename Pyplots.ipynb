```python
# Create a DataFrame
data = pd.DataFrame({
    'Input': np.linspace(0, 10, 100),
    'Output_1': np.sin(np.linspace(0, 2*np.pi, 100)),
    'Output_2': np.cos(np.linspace(0, 2*np.pi, 100))
})

# Plot the data
ax = data.plot(figsize=(8, 6))

# Customize using the Axes 
ax.grid(True)
ax.set_title('Line Plot') 
ax.set_xlabel('X Values')  
ax.set_ylabel('Y Values')  
ax.legend(loc='lower left') 

# Show the plot
plt.show()
```

# Line Plot
The default plot for Pandas is a line plot.

```python
data.plot(kind='line', figsize=(8, 6), title="Line Plot", legend=True, grid=True)
```

# Bar Plot

Used to visualize data in vertical bars.

```python
data.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'orange', 'green'], title="Bar Plot", rot=0)
```

```python
data.plot(kind='bar', stacked=True, figsize=(8, 6), title="Stacked Bar Plot")
```

# Horizontal Bar Plot

Flip the bars horizontally.

```python
data.plot(kind='barh', figsize=(8, 6), color=['green', 'purple', 'red'], title="Horizontal Bar Plot")
```

# Histogram

Used to visualize the distribution of a single column or multiple columns.

## Parameters:
    bins: Number of bins in the histogram.
    alpha: Transparency level (0.0 is fully transparent, 1.0 is opaque).

```python
data['Output_1'].plot(kind='hist', bins=20, color='teal', alpha=0.7, figsize=(8, 6), title="Histogram")
```

# Box Plot

Used to display the spread and outliers in the data.

```python
data.plot(kind='box', figsize=(8, 6), title="Box Plot")
```

# Scatter Plot

Used to visualize relationships between two columns.

```python
data.plot(kind='scatter', x='Input', y='Output_1', color='blue', figsize=(8, 6), title="Scatter Plot")
```

```python
data.plot(kind='scatter', x='Input', y='Output_1', c='Output_2', colormap='viridis', figsize=(8, 6))
```

# Area Plot

Used for cumulative visualizations.

```python
data.plot(kind='area', stacked=False, alpha=0.5, figsize=(8, 6), title="Area Plot")
```

# Pie Chart

Used for proportions in a single column.

## Parameters:
    autopct: Display percentages on the chart.
    startangle: Rotate the start angle of the pie.

```python
data['Output_1'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title="Pie Chart")
```

# Hexbin Plot

Used for density visualization in large scatter plots.

```python
data.plot(kind='hexbin', x='Input', y='Output_1', gridsize=25, colormap='Blues', figsize=(8, 6), title="Hexbin Plot")
```

# KDE (Kernel Density Estimation) Plot

Used to visualize the probability density of a dataset.

## Parameters:
    bw_method: Adjust the bandwidth of the kernel.

```python
data['Output_1'].plot(kind='kde', figsize=(8, 6), color='red', title="KDE Plot")
```
